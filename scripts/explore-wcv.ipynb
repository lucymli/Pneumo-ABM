```python
from collections import namedtuple
from functools import partial
import palettable
import json
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
from plotting_utils import remove_spines, save_figure
```

```python
%matplotlib inline
COLORS = palettable.colorbrewer.qualitative.Set1_9.mpl_colors
```

```python
RUN_NAME = '0.8-sr-only'
NUM_TRIALS = 1
OUTPUT_DIR = '../outputs/INCIDENCE/{}/'.format(RUN_NAME)
```

```python
save_fig = partial(save_figure, folder=OUTPUT_DIR + 'figures', exts=['pdf', 'png'])
```

```python
with open(OUTPUT_DIR + 'configuration/configuration.json') as f:
    CONFIG = json.load(f, object_hook=lambda d: namedtuple('Config', d.keys())(*d.values()))
SEROTYPES = json.load(open(OUTPUT_DIR + 'configuration/' + CONFIG.serotype.serotypes_file))['serotypes']
NUM_SEROTYPES = len(SEROTYPES)
MAX_AGE = CONFIG.population.max_age
NUM_YEARS = CONFIG.simulation.num_years_burn_in.demographic + CONFIG.simulation.num_years_burn_in.epidemiologic + CONFIG.simulation.num_years_simulation + 1
VACCINATION_START = CONFIG.simulation.vaccinations[0].start_year
VACCINE = CONFIG.simulation.vaccinations[0].vaccine
NUM_HOSTS = CONFIG.simulation.num_initial_hosts
```

```python
NUM_HOSTS
```

```python
HI_PREV_SEROTYPES = ('19F', '6A', '6B', '23F', '11A')
```

```python
def read_csv(filename):
  return np.array(pd.read_csv(filename, header=None))
```

### Plot Serotypes

```python
get_num_colonized_st = lambda t, st: read_csv('{}trial-{}/num_colonized_by_age_ss_{}.csv'.format(OUTPUT_DIR, t, st)).sum(axis=1)
get_num_colonized    = lambda t: read_csv('{}trial-{}/num_colonized.csv'.format(OUTPUT_DIR, t)).flatten()
get_num_colonized_u5 = lambda t: read_csv('{}trial-{}/num_colonized_under_5.csv'.format(OUTPUT_DIR, t)).flatten()
get_num_hosts        = lambda t: read_csv('{}trial-{}/num_hosts.csv'.format(OUTPUT_DIR, t)).flatten().astype(float)
get_num_hosts_u5     = lambda t: read_csv('{}trial-{}/num_hosts_by_age.csv'.format(OUTPUT_DIR, t))[:,0:5].sum(axis=1).astype(float)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
for i, st in enumerate(HI_PREV_SEROTYPES):
  for t in range(NUM_TRIALS):
    ys = get_num_colonized_st(t, st) / get_num_hosts(t)
    ax.plot(ys, alpha=0.6, c=COLORS[i % len(COLORS)], label=st if t==0 else None)

ax.axvline(x=VACCINATION_START, c='k', ls='--', lw=1)
ax.set_xlim(VACCINATION_START - 50)
remove_spines(ax)
ax.legend(frameon=True)
ax.set_ylabel('Prevalence')
ax.set_xlabel('Simulation Year')
ax.grid(True, alpha=0.5)
ax.set_title(RUN_NAME, y=1.02)
fig.tight_layout()
save_fig(fig, "serotype-dynamics-{}".format(RUN_NAME))
```

### Plot Overall Prevalence in Kids and Adults

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
for t in range(NUM_TRIALS):
  total_prev = get_num_colonized(t) / get_num_hosts(t)
  u5_prev = get_num_colonized_u5(t) / get_num_hosts_u5(t)
  ax.plot(total_prev, c='k', alpha=0.5, label='all ages' if t == 0 else None)
  ax.plot(u5_prev, c=COLORS[7], alpha=0.8, label='< 5 yrs old' if t == 0 else None)

ax.axvline(x=VACCINATION_START, c='k', ls='--', lw=1)

ax.set_ylim(0, 0.8)
ax.set_xlim(VACCINATION_START - 50)
ax.set_ylabel("Prevalence")
ax.set_xlabel("Simulation Year")
ax.legend(loc='lower left', frameon=True)

ax.grid(True)
ax.xaxis.set_major_locator(mpl.ticker.MultipleLocator(10))
ax.set_title(RUN_NAME, y=1.02)

remove_spines(ax)
fig.tight_layout()
save_fig(fig, "prevalence-dynamics-{}".format(RUN_NAME))
```

### Table of Serotype Prevalences

```python
def get_avg_prev(serotype, start, years):
  avg_by_trial = []
  for t in range(NUM_TRIALS):
    prev = get_num_colonized_st(t, serotype) / get_num_hosts(t)
    avg_by_trial.append(prev[start:start + years].mean())
  return np.array(avg_by_trial).mean()
```

```python
%%time
table = pd.DataFrame(index=SEROTYPES)
table['prevaccine'] = pd.Series()
table['postvaccine'] = pd.Series()
for st in SEROTYPES:
  table.loc[st, 'prevaccine'] = get_avg_prev(st, VACCINATION_START - 10, 10)
  table.loc[st, 'postvaccine'] = get_avg_prev(st, VACCINATION_START + 5, 10)
```

```python
table.sort('prevaccine', inplace=True, ascending=False)
```

```python
table['percent-decrease'] = 100 * (table.prevaccine - table.postvaccine)  / table.prevaccine
```

```python
with open(OUTPUT_DIR + 'figures/{}.csv'.format(RUN_NAME), 'w') as f:
  table.to_csv(f, float_format="%.6f")
```

### Plot Incidence

```python
def get_num_incidents(t, group, st='all'):
  if st=='all': 
    return read_csv('{}trial-{}/incidence/incidence_{}.csv'.format(OUTPUT_DIR, t, group)).sum(axis=1)
  else:
    return read_csv('{}trial-{}/incidence/incidence_{}.csv'.format(OUTPUT_DIR, t, group))[:,serotypes.index(st)]
```

```python
get_num_vaccinated = lambda t: read_csv('{}trial-{}/num_vaccinated_all_ages_WCV.csv'.format(OUTPUT_DIR, t)).flatten()
```

```python
get_host_years = lambda ts: (ts + np.append([0], ts[:-1])) / 2.0
```

#### Raw numbers

```python
fig = plt.figure(figsize=(12, 10))
gs = mpl.gridspec.GridSpec(3, 1, height_ratios=[5, 2, 5]) 
axes = [plt.subplot(gs[i]) for i in range(3)]

colors = (COLORS[1], COLORS[0], 'k')
groups = ("vaccinated", "unvaccinated", "all")
ax = axes[0]
for t in range(NUM_TRIALS):
  #sp = ax.stackplot(np.arange(len(y1)), y1, y2, colors=(COLORS[1], COLORS[0]), alpha=1.0, color='none')
  #proxy = [mpl.patches.Rectangle((0,0), 0,0, facecolor=pol.get_facecolor()[0]) for pol in sp]
  for c, g in zip(colors, groups):
    ax.plot(get_num_incidents(t, g), color=c, label="in {} hosts".format(g))

ax.axvline(x=VACCINATION_START, c='k', ls='--', lw=1)
ax.legend(loc='lower left', frameon=True)
#ax.legend(proxy, ('in vaccinated hosts', 'in unvaccinated hosts'), loc='lower left', frameon=True)
ax.set_xlim(VACCINATION_START - 50)
ax.set_ylabel("Colonizations per Year")
ax.set_title('Annual Number of Colonizations')


ax = axes[1]
for t in range(NUM_TRIALS):
  ax.plot(get_num_vaccinated(t) / float(NUM_HOSTS), lw=1.5, ls='-', c='k', label='coverage' if t == 0 else None)

ax.axvline(x=VACCINATION_START, c='k', ls='--', lw=1)

ax.legend(loc='lower left', frameon=True)
ax.set_ylabel("Coverage")
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.set_xlabel("Simulation Year")
ax.set_title('Vaccine Coverage')

ax = axes[2]
for t in range(NUM_TRIALS):
  y0 = get_num_incidents(t, 'vaccinated') / get_host_years(get_num_vaccinated(t))
  y1 = get_num_incidents(t, 'unvaccinated') / get_host_years(NUM_HOSTS - get_num_vaccinated(t))
  y2 = get_num_incidents(t, 'all') / float(NUM_HOSTS)

  ax.plot(y0, color=COLORS[1], label='in vaccinated hosts')
  ax.plot(y1, color=COLORS[0], label='in unvaccinated hosts')
  ax.plot(y2, color='k', label='in all hosts')

ax.axvline(x=VACCINATION_START, c='k', ls='--', lw=1)
legend = ax.legend(loc='lower left', frameon=True)

ax.set_xlim(VACCINATION_START - 50)
ax.set_ylabel("Colonizations per Host-Year")
ax.set_title('Incidence Rate')

for a in axes:
  a.set_xlim(VACCINATION_START - 50)
  a.grid(True, alpha=0.5)
  a.xaxis.set_major_locator(mpl.ticker.MultipleLocator(10))
  remove_spines(a)

fig.tight_layout()
fig.suptitle(RUN_NAME, y=1.03, fontsize=18)
save_fig(fig, "annual-incidence-{}".format(RUN_NAME))
```
