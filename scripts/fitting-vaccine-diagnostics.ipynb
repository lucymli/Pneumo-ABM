### Compare Estimated Prevalences and Target Prevalences

```python
vaccine = filter(lambda v: v.name == config.fitting.vaccine_to_fit, config.vaccines)[0]
```

```python
is_vaccine_type = np.array([s in set(vaccine.susceptibility_reduction.serotypes) for s in serotype_names])
```

```python
estimated_vt_prev = np.sum(estimated_prevs[:,:-1] * is_vaccine_type, axis=1)
target_vt_prev    = np.sum(target_prevs[:-1] * is_vaccine_type)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
ax.plot(estimated_vt_prev)
ax.axhline(y=target_vt_prev, c='.5', ls='--')
remove_spines(ax)
ax.set_ylabel('Vaccine Type Prevalence')
ax.set_xlabel('Iteration')
save_fig(fig, 'total-prevalence')
```

Compute a moving average

```python
def moving_average(array, n):
    S = np.cumsum(array, dtype=float)
    S[n:] = (S[n:] - S[:-n]) / n        # average most recent n elements
    S[:n] = S[:n] / np.arange(1, n + 1) # average available elements for the first n windows
    return S
```

```python
window_size = 15
estimated_prevs_mva = np.zeros_like(estimated_prevs)
for i in range(estimated_prevs.shape[1]):
    estimated_prevs_mva[:, i] = moving_average(estimated_prevs[:, i], n=window_size)
```

Look at estimates from each iteration

```python
num_cols = 4
num_rows = num_serotypes / num_cols
fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 3, num_rows * 2))

for i, ax in enumerate(axes.flatten()):
    if i > num_serotypes - 1:
        ax.axis('off')
        continue

    xs = np.arange(0, num_iterations)
    ys = estimated_prevs[:,i]
    row = i / num_cols

    ax.set_xlim(0, max(xs) + 1)
    ax.set_ylim(0, max(max(ys), target_prevs[i]) * 1.01)

    a = 1 if is_vaccine_type[i] else 0.5
    ax.scatter(xs, ys, c=st2color[i], alpha=a, s=20)
    ax.axhline(y=target_prevs[i], c='k', ls='--', lw=2, alpha=a)
    ax.plot(xs, estimated_prevs_mva[:, i], c='k', lw=1.5, alpha=a)
    
    fw = 'bold' if is_vaccine_type[i] else 'normal'
    ax.text(num_iterations / 2, ax.get_ylim()[1] * 0.95, '{} : {:.2f}'.format(serotype_names[i], ranks[i]), va='top', ha='center', 
            fontsize=18, fontweight=fw, color='.3')
    
    ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
    ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))    
    
    remove_spines(ax)
    
fig.tight_layout()
save_fig(fig, 'convergence-window-size-{}'.format(window_size))
```
